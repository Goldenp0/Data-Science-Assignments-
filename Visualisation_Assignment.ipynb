```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('birds.csv')

# Filter data for Snow Geese
snow_geese_data = df[df['Name'] == 'scientificname']

# Plot 1: Line Chart - Population over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Name', y='ScientificName', data=snow_geese_data, marker='o')
plt.title('Bird Trend')
plt.xlabel('Name')
plt.ylabel('ScientificName')
plt.grid(True)
plt.show()

# Plot 2: Scatterplot - Relationship between Two Variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Name', y='ScientificName', data=snow_geese_data, hue='Category', palette='viridis', s=100)
plt.title('Bird family')
plt.xlabel('Name')
plt.ylabel('ScientificName')
plt.legend(title='Category')
plt.grid(True)
plt.show()

# Plot 3: Bar Chart - Comparison across Categories
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Order', data=snow_geese_data, palette='coolwarm')
plt.title('Comparison of Categories in birds')
plt.xlabel('Category')
plt.ylabel('Order')
plt.grid(axis='y')
plt.show()
```
